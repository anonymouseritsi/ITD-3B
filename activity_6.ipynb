```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

```python


# Load the dataset with the correct encoding
df = pd.read_csv('datasets\\Global_terrorist.csv', encoding='ISO-8859-1')

```

```python
df
```

```python
import folium
```

<H1>Filtering Data</H1>

```python
df['region_txt'] == 'Middle East & North Africa'
```

```python
filter1 = df[df['region_txt'] == 'Middle East & North Africa']
```

```python
filter1['iyear'] > 2000
```

```python
filter2 = filter1[filter1['iyear'] > 2000]
```

```python
filter2['attacktype1_txt'].value_counts()
```

```python
filter2['attacktype1_txt'] == 'Bombing/Explosion'
```

```python
filter3 = filter2[filter2['attacktype1_txt'] == 'Bombing/Explosion']
```

```python
long_lat_only = filter3.groupby(['latitude','longitude']).size().reset_index(name='incident_count')
```

```python
long_lat_only
```

<H1>Use of FOLIUM FOR GEOSPATIAL</H1>

```python
from folium.plugins import HeatMap
```

<h1>Converting PANDAS df into Python List</h1>

```python
long_lat_only['normalize_data'] = long_lat_only['incident_count'] - long_lat_only['incident_count'].min() / long_lat_only['incident_count'].max() - long_lat_only['incident_count'].min()
heat_data = long_lat_only[['latitude','longitude','normalize_data']].values.tolist()
```

<h1>Creating a base map</h1>

```python
my_loc = folium.Map(location =[30,40],zoom_start =4)
```

<H1>Add Heatmap Layer</H1>

```python
HeatMap(heat_data).add_to(my_loc)
```

```python
my_loc.save('terrorism_bombing.html')
print("Data Successfully Saved to terrorism_bombing.html")
```
