<H1>DOWNLOADING LIBRARY (SCIPY) ONE WAY ANOVA</H1>

```python
pip install scipy
```

```python
from scipy.stats import f_oneway
```

```python
penguins[penguins['sex'] == 'male']['body_mass_g']
```

```python
penguins[penguins['sex'] == 'female']['body_mass_g']
```

```python
f_stats,p_value = f_oneway(penguins[penguins['sex'] == 'male']['body_mass_g'],
                    penguins[penguins['sex'] == 'female']['body_mass_g'])
print(p_value)
```

<h1>---Insight 16. There is a correlation between sex and body mass (ONE WAY ANOVA)</h1>
