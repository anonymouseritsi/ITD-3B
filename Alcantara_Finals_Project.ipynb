```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('datasets\\Family_Income_and_Expenditure.csv')
```

```python
df.columns
```

```python
df.dtypes
```

```python
df['Region'] = df['Region'].astype('category')
df['Main Source of Income'] = df['Main Source of Income'].astype('category')
df['Household Head Sex'] = df['Household Head Sex'].astype('category')
df['Household Head Marital Status'] = df['Household Head Marital Status'].astype('category')
df['Household Head Highest Grade Completed'] = df['Household Head Highest Grade Completed'].astype('category')
df['Household Head Job or Business Indicator'] = df['Household Head Job or Business Indicator'].astype('category')
df['Household Head Occupation'] = df['Household Head Occupation'].astype('category')
df['Household Head Class of Worker'] = df['Household Head Class of Worker'].astype('category')
df['Type of Building/House'] = df['Type of Building/House'].astype('category')
df['Tenure Status'] = df['Tenure Status'].astype('category')
df['Type of Household'] = df['Type of Household'].astype('category')
df['Main Source of Water Supply'] = df['Main Source of Water Supply'].astype('category')

```

```python
df['Non-food Expenditure'] = np.round(df['Total Household Income']-df['Total Food Expenditure'],1)
```

```python
df
```

```python
car = df[df['Region'] == 'CAR']
caraga = df[df['Region'] == 'Caraga']
I = df[df['Region'] == 'I - Ilocos Region']
II = df[df['Region'] == 'II - Cagayan Valley']
III = df[df['Region'] == 'III - Central Luzon']
IVA = df[df['Region'] == 'IVA - CALABARZON']
IVB = df[df['Region'] == 'IVB - MIMAROPA']
V = df[df['Region'] == 'V - Bicol Region']
VI = df[df['Region'] == 'VI - Western Visayas']
VII = df[df['Region'] == 'VII - Central Visayas']
VIII = df[df['Region'] == 'VIII - Eastern Visayas']
IX = df[df['Region'] == 'IX - Zasmboanga Peninsula']
X = df[df['Region'] == 'X - Northern Mindanao']
XI = df[df['Region'] == 'XI - Davao Region']
XII = df[df['Region'] == 'XII - SOCCSKSARGEN']
ncr = df[df['Region'] == 'NCR']
```

```python
regionfood = df.groupby(['Region'])['Total Food Expenditure'].sum()
```

```python
regionfood
```

```python
region_non_food = df.groupby(['Region'])['Non-food Expenditure'].sum()
```

```python
region_non_food
```

```python
regiontotal =df.groupby(['Region'])['Total Household Income'].sum()
```

```python
regiontotal
```

```python
percent =  np.round(regionfood/regiontotal,2)
```

```python
percent.reset_index(name='percentage')
```

```python
percent.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Percentage Rate of Total Food Expenditure over Total Income Per Region') 
plt.show()
```

<h1>Insight 1.</h1>
<h2>Regions like ARMM (0.48) and Bicol (0.41) have the highest share of household income spent on food, indicating lower income levels and potential food insecurity. In contrast, regions like NCR, CAR, and Northern Mindanao (all at 0.30) spend a smaller proportion, suggesting higher income and better economic conditions.</h2>

```python
percent_nonfood =  np.round(region_non_food/regiontotal,2)
```

```python
percent_nonfood.reset_index()
```

```python
percent_nonfood.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Percentage Rate of Total Non Food Expenditure over Total Income Per Region') 
plt.show()
```
