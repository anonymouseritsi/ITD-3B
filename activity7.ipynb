```python
battery['Location Description'].value_counts()
```

```python
theft_loc = theft[theft['Location Description']=='APARTMENT']
battery_loc = battery[battery['Location Description']=='STREET']
```

```python
battery_loc['Year'].value_counts()
```

```python
year_tl = theft_loc[theft_loc['Year']==2025]
year_bl = battery_loc[battery_loc['Year']==2025]
```

```python
theft_final = year_tl.groupby(['Latitude','Longitude']).size().reset_index(name='incident_count')
battery_final = year_bl.groupby(['Latitude','Longitude']).size().reset_index(name='incident_count')
```

```python
theft_final
```

```python
theft_final['normalize_data'] = theft_final['incident_count'] - theft_final['incident_count'].min() / theft_final['incident_count'].max() - theft_final['incident_count'].min()
theft_data = theft_final[['Latitude','Longitude','normalize_data']].values.tolist()

battery_final['normalize_data'] = battery_final['incident_count'] - battery_final['incident_count'].min() / battery_final['incident_count'].max() - battery_final['incident_count'].min()
battery_data = battery_final[['Latitude','Longitude','normalize_data']].values.tolist()
```

```python
theft_map = folium.Map(location=[41.8781, -87.6298], zoom_start=10)
```

```python
HeatMap(theft_data).add_to(theft_map)
```

```python
theft_map.save('chicago_crimes.html')
```
